```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pathToProject = 'D:\Work\WorkSpace\CLionProjects\VKR-1\cmake-build-debug\\'
files = [
    'chebyshev_long_double_sin.csv', 
    'chebyshev_double_sin.csv', 
    
    'legendre_long_double_sin.csv', 
    'legendre_double_sin.csv', 
    
    'taylor_long_double_sin.csv',
    'taylor_double_sin.csv'
         ]
data_example = pd.read_csv(pathToProject + 'chebyshev_long_double_sin.csv', delimiter=';')
x_value = data_example['x'][0]
sin_ = np.sin(0.5)
```

```python
def createPlt():
    plt.figure(figsize=(14, 6))

    for file in files:
        data = pd.read_csv(pathToProject + file, delimiter=';')
        plt.plot(data['step'], data['result'], label=file[:-4])

    plt.axhline(y=sin_, color='g', linestyle='--', label=f'sin = {sin_:.10f}')

    plt.xlabel('Step')
    plt.ylabel('Result')
    plt.title('График данных')
    plt.grid(True)
    
    return plt
```

```python
plt = createPlt()
plt.legend(loc='upper right')
plt.show()

plt = createPlt()
plt.ylim(1e-19, 1e-2)
plt.legend(loc='upper right')
plt.yscale('log')
plt.show()
```
